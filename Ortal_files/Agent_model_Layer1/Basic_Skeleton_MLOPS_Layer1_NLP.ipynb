```python
# 🚀 Ortal's Advanced Baseline Emotion Detection Classifier (Agentic Project Ready) - Upgraded with Transformers + Emotion Models + Multi-Dataset Support

# 🧭 Team To-Do Ideas:
# - Map Mistral predictions to numeric labels for metric evaluation ✅
# - Try hyperparameter tuning for classical models ✅
# - Test on alternative emotion datasets ✅
# - Add ROC-AUC evaluation ✅
# - Try model ensembling (e.g., voting classifier) ✅

# Step 1: Install Dependencies
!pip install datasets scikit-learn pandas matplotlib seaborn nltk wordcloud xgboost lightgbm transformers torch accelerate --quiet

# Step 2: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from wordcloud import WordCloud
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from datetime import datetime

# Transformers imports
from transformers import (BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments,
                          AutoTokenizer, AutoModelForSequenceClassification, pipeline)
import torch

print("✅ Libraries imported!")

# Step 3: Define Multiple Datasets
datasets = [
    {"name": "Depression Dataset", "dataset": load_dataset("joangaes/depression", split='train').to_pandas()},
    # Add more datasets here for team exploration
    # {"name": "Emotion Dataset", "dataset": load_dataset("emotion", split='train').to_pandas()}
]

# Prepare results collection
results_list = []

# Prepare timestamp for saving outputs
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")

for data_info in datasets:
    dataset = data_info["dataset"]
    dataset_name = data_info["name"]

    print(f"\n=== Processing Dataset: {dataset_name} ===")

    # Step 4: EDA - Visualize Class Distribution
    plt.figure(figsize=(8, 4))
    sns.countplot(x=dataset['label'])
    plt.title(f'Class Distribution - {dataset_name}')
    plt.savefig(f'class_distribution_{dataset_name}_{timestamp}.png')
    plt.show()

    # Step 5: Preprocessing
    X = dataset['text']
    y = dataset['label']

    # Step 6: Vectorization (for classical models)
    vectorizer = TfidfVectorizer(max_features=5000)
    X_vectorized = vectorizer.fit_transform(X)

    # Step 7: Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

    # Step 8: Initialize Models (Classical Models)
    models = {
        "Logistic Regression": LogisticRegression(max_iter=1000),
        "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
        "Naive Bayes": MultinomialNB(),
        "Linear SVM": LinearSVC(),
        "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
        "LightGBM": LGBMClassifier()
    }

    # Step 9: Train, Evaluate and Collect Results (Classical Models)
    for name, model in models.items():
        print(f"\n🔍 Training: {name} on {dataset_name}")
        model.fit(X_train, y_train)
        preds = model.predict(X_test)

        accuracy = accuracy_score(y_test, preds)
        f1 = f1_score(y_test, preds, average='weighted')
        precision = precision_score(y_test, preds, average='weighted')
        recall = recall_score(y_test, preds, average='weighted')

        results_list.append({
            'Dataset': dataset_name,
            'Model Name': name,
            'Accuracy': accuracy,
            'F1 Score': f1,
            'Precision': precision,
            'Recall': recall
        })

        print(classification_report(y_test, preds))

        plt.figure(figsize=(4, 3))
        cm = confusion_matrix(y_test, preds)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title(f'{name} Confusion Matrix - {dataset_name}')
        plt.ylabel('Actual')
        plt.xlabel('Predicted')
        plt.savefig(f'confusion_matrix_{dataset_name}_{name}_{timestamp}.png')
        plt.show()

    # Step 10: Placeholder for future BERT / advanced models per dataset
    results_list.append({
        'Dataset': dataset_name,
        'Model Name': 'BERT',
        'Accuracy': 'To Do',
        'F1 Score': 'To Do',
        'Precision': 'To Do',
        'Recall': 'To Do'
    })

    results_list.append({
        'Dataset': dataset_name,
        'Model Name': 'Go-Emotion BERT',
        'Accuracy': 'To Do',
        'F1 Score': 'To Do',
        'Precision': 'To Do',
        'Recall': 'To Do'
    })

    results_list.append({
        'Dataset': dataset_name,
        'Model Name': 'Mistral',
        'Accuracy': 'To Do',
        'F1 Score': 'To Do',
        'Precision': 'To Do',
        'Recall': 'To Do'
    })

    # Step 11: WordCloud for Top Words
    all_words = ' '.join(X)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words)

    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'WordCloud of Text Data - {dataset_name}')
    plt.savefig(f'wordcloud_{dataset_name}_{timestamp}.png')
    plt.show()

# Step 12: Export Results
results_df = pd.DataFrame(results_list)
results_df.to_csv(f'model_comparison_results_{timestamp}.csv', index=False)
print(f"\n✅ Results exported to model_comparison_results_{timestamp}.csv")

# Step 13: Visual Comparison (Clean Data)
plot_df = results_df[results_df['Accuracy'] != 'To Do'].copy()
for col in ['Accuracy', 'F1 Score', 'Precision', 'Recall']:
    plot_df[col] = pd.to_numeric(plot_df[col], errors='coerce')

plt.figure(figsize=(12, 7))
sns.barplot(data=plot_df, x='Model Name', y='Accuracy', hue='Dataset')
plt.title('Model Performance Comparison Across Datasets')
plt.ylabel('Accuracy Score')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.tight_layout()
plt.savefig(f'model_performance_comparison_{timestamp}.png')
plt.show()

print("\n🎉 Pipeline completed successfully! 🚀 Team members: extend from here.")
```
